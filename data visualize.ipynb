```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x =[2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018]
y =[205.9,208.6,212.7,220.5,233.2,242.4,245.4,241.4,250.2,255.7,261.2,267.9,280.1]
z =[11402,11492,11450,11502,11474,11480,11452,11484,11555,11601,11731,11862,12048]


plt.bar(x=x,height=y,label="WO")
plt.legend(loc='upper right')

plt.title("graph")
plt.xlabel("Years")
plt.ylabel("WO[x1000]")

ax2 = plt.twinx()
ax2.set_ylabel("Beer[x1000]")
plt.plot(x,z,"r",label="Beer")

plt.legend(loc='upper left')
plt.savefig('Beer.jpg')

plt.show()
```
